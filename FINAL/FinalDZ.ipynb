# Исходные данные

```python
from sympy import *
```

```python
x = Symbol('x')
```

```python
func = (sin(x)/2 + cos(x)/2) ** 2 * sin(x)
func
```

# Корни

```python
coren = sorted(solve(func))
coren
```

# Интервалы возрастания / убывания

```python
proi = diff(func)
proi
```

```python
corproi = sorted(solve(proi))
corproi
```

```python
for i in range(len(corproi)-1):
    y = proi.subs(x, (corproi[i]+corproi[i+1])/2) 
    if y > 0:
        print(f'Функция убывает от {corproi[i].evalf()} до {corproi[i+1].evalf()}')
    else:
        print(f'Функция возрастаяет от {corproi[i].evalf()} до {corproi[i+1].evalf()}')      
```

# График

```python
gr = plot(func, proi, (x,-pi,pi))
```

# Вершины

```python
for i in corproi:
    y = func.subs(x, i)
    print('x = ', i.evalf(), '| y = ', y.evalf())
```

# Промежутки 0<f<0

```python
solve(func > 0)
```

```python
solve(func < 0)
```
