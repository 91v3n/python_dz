# Исходные данные

```python
from sympy import *
```

```python
x = Symbol('x')
```

```python
func = (sin(x)/2 + cos(x)/2) ** 2 * sin(x)
func
```

# Корни

```python
coren = sorted(solve(func))
coren
```

# Интервалы возрастания / убывания

```python
proi = diff(func)
proi
```

```python
corproi = sorted(solve(proi))
corproi
```

```python
half = periodicity(func, x)/2
half
```

```python
ext = [i for i in corproi]
ext.append(half)
ext.insert(0, -half)
```

```python
for i in range(len(ext)-1):
    y = proi.subs(x, (ext[i]+ext[i+1])/2) 
    if y < 0:
        print(f'Функция убывает от {ext[i].evalf()} до {ext[i+1].evalf()}')
    else:
        print(f'Функция возрастаяет от {ext[i].evalf()} до {ext[i+1].evalf()}')      
```

# График

```python
gr = plot(func, proi, (x,-half,half))
```

# Вершины

```python
for i in corproi:
    y = func.subs(x, i)
    print('x = ', i.evalf(), '| y = ', y.evalf())
```
